## Pyramid of DOOM

Zu viele Verschachtelungen machen code schwer zu lesen. Wir versuchen daher code so zu schreiben, dass wir Verschachtelung vermeiden. Das ist nicht immer möglich aber oft kann es mit einer kleinen Umstrukturierung vermieden werden.

Pyrymid of DOOM kann in jeder Sprache entstehen. Hier ein beispiel in Java.

pyramid.png


Verschachtelung (nesting) kann auftreten bei:
`if else`, `while`, `for`, `def` ... immer wenn wir eine Einrückung benötigen.

Hier haben wir einen kleinen Code mit Verschachtelung. Dieses Beispiel soll zeigen wie schlechter Code aussieht. Der Code ist zwar korrekt, aber nicht "schön".


````python
import random
````

````python
guesses = 5
min_num = 1
max_num = 10
answer = random.randint(min_num,max_num)
print(f"I'm thinking about a number between {min_num} and {max_num}")

while(True): # Endless loop, only stops on break.
    if guesses > 0:
        guess = float(input('Guess a number: '))
        guesses-=1

    if guess == answer:
        print("You win")
        break
    elif guess < answer:
        print("Too low")
    elif guess > answer:
        print("Too high")
    else:
        print("You lose")
        break
````

Es gib eine Strategie die fast immer funktioniert:

## Guard Clauses

Wir schreiben die Bedingungen die die Ausführung beenden möglichst früh. Nur wenn diese "Wächter" passiert werden geht es weiter, sonst beenden wir die Ausführung mit `break`, `return`, `raise`, ...

Es gibt viele namen für diese Strategie: *Early return*, *Guard Clauses*, *Special Cases First*, ...

````python
guesses = 5
min_num = 1
max_num = 10
answer = random.randint(min_num,max_num)
print(f"I'm thinking about a number between {min_num} and {max_num}")

while(True): # Endless loop, only stops on break.
    if guesses == 0:
        print("You lose")
        break

    guess=int(input('Guess a number: '))
    guesses-=1

    if guess == answer:
        print("You win")
        break

    if guess < answer:
        print("Too low")
    else:
        print("Too high")
````

Das Problem ist nicht Performance, sondern Lesbarkeit.
Performance ist bei Programmen meistens nicht das Problem.

Code wird in der Regel 10x öfter gelesen als geschrieben. 

- In einem professionellem Umfeld wird Code von mindestens einer Person angesehen bevor er in ein System übernommen wird.
- Code wird verändert. Anforderungen ändern sich. Code muss nach Wochen oder Jahren wieder gelesen werden um zu verstehen was er macht.
- Code hat fehler. Größere Projekte haben nicht 20 Zeilen, sondern Tausende oder sogar Millionen. Niemand kennt da noch jede Zeile, und Fehler passieren auch im Zusammenspeil von kleineren Teilen im Code.

Es ist also hilfreich Code einfach zu schreiben. Pyramids of Doom machen es schwer Code zu lesen und zu verstehen.

## Funktionen

Funktionen sind ein einfaches Mittel um Code zu strukturieren und wiederzuverwenden.

Funktionen werden mit dem Keyword `def` definiert.
Danach bekommt die Funktion einen Namen. Nach dem Namen kommen die Parameter der Funktion in Klammern `()` gefolgt von einem `:`.

```python
def name_der_funktion(parameter_1, parameter_2):
  pass
  ```

`pass` bedeutet hier dass die Funktion nichts macht.



````python
def double(number):
    return number * 2
````

Wir haben hier die Funktion `double` definiert.
Double bekommt eine Zahl `number` und gibt den doppelten wert zurück.

Die Definition eine Funktion alleine macht noch nichts, wir müssen die Funktion erst aufrufen. 

````python
double(21)
````

````python
def square(number):
    return number ** 2
````

````python
square(7)
````

Der rückgabe wert dieser Funktionen ist vom Type `float` oder `int`, und kann selbst wieder verwendet werden.


````python
double(square(3)) # 3*3*2
````

````python
square(double(3)) # (3*2)*(3*2)
````

Funktionen sollten immer möglichst wenig machen. Eine Funktion sollte eine Aufgabe haben.

Aber es ist auch ok Funktionen zu haben die andere Funktionen kombinieren.

````python
def square_of_double(number):
    return square(double(number))
````

````python
square_of_double(3) # (3*2)*(3*2)
````

`square_of_double` ruft zuerst `double` auf und verwendet das Ergbnis in `square`

Funktionen können sich auch selbst wieder aufrufen, aber Vorsicht (!) es ist wichtig zu verhindern, dass die Funktion sich selbst unendlich oft aufruft.

solche Funktionen nennt man rekursive Funktionen.

````python
def rec_bad():
    rec_bad()
````

````python
rec_bad()
````

`rec_bad` ist eine böse rekursive Funktion. Sie ruft sich selbst solange auf bis der Pythoninterpreter keinen Speicherlatz mehr hat um die Funktionsaufrufe zu verwalten.

````python
def little_gauss(num):
    if num <= 0:
        #return -1 nicht gut
        raise ValueError('No negative numbers')
    if num == 1:
        return 1
    else:
        return num + little_gauss(num - 1)
````

````python
little_gauss(-1)
````

`little_gauss` addiert alle positven Zahlen von `1` bis `num`.
```
Die Summe von 1 bis 5, ist gleich wie 5 + die Summe von 1 bis 4.
Die Summe von 1 bis 4, ist gleich wie 5 + die Summe von 1 bis 3.
.
.
.
Die Summe von 1 bis 1 ist 1.
```
Für 1 haben wir einen speziellen Fall definiert, dadurch gibt es keine unendliche Rekursion. Wir sagen, dass die Funktion **terminiert** also aufhört. num == 1 wird oft auch Abbruchsbedingung oder Randfall genannt.
